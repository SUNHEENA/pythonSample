```python
# 네이버 
import requests # 사이트의 webpage 문서를 back ground 에서 down할 수 있도록
from bs4 import BeautifulSoup
ent = input("기업을 입력 해 주세요-->")
url = "https://search.naver.com/search.naver?where=nexearch&sm=top_hty&fbm=1&ie=utf8&query=" + ent + " 주가"
html = requests.get(url).text
soup = BeautifulSoup(html, 'html.parser')
tempTxt = soup.find('span', 
        class_='spt_con up').text
if tempTxt =='':
    soup.find('span', 
        class_='spt_con eq').text
t = tempTxt.split(' ')
print(f"현재 {ent} 주가는 {t[2]} 입니다.")
```

```python
#find_elements_by_xpath
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By

driver = webdriver.Chrome(ChromeDriverManager().install())
driver.get("https://www.melon.com/chart/")

title_list = driver.find_elements(By.XPATH,"//div[contains(@class,'ellipsis rank01')]")
singer_list = driver.find_elements(By.XPATH,"//div[contains(@class,'ellipsis rank02')]")
for i in range(0, len(title_list)):
    print("가수:"+singer_list[i].text+", 제목 :"+title_list[i].text)
driver.quit()
```

```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
driver = webdriver.Chrome(ChromeDriverManager().install())
driver.get("https://www.melon.com/chart/")

title_list = driver.find_elements_by_xpath("//div[contains(@class,'ellipsis rank01')]")
singer_list = driver.find_elements_by_xpath("//div[contains(@class,'ellipsis rank02')]")
titleList=list()
nameList=list()
for i in range(0, len(title_list)):
    #print("가수:"+singer_list[i].text+", 제목 :"+title_list[i].text)
    titleList.append(title_list[i].text)
    nameList.append(singer_list[i].text)
driver.quit()
data = {
        '제목': titleList,
        '가수': nameList
        }
df = pd.DataFrame(data,columns=['제목','가수'])
df
```

```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
import win32com.client

driver = webdriver.Chrome(ChromeDriverManager().install())
driver.get("https://www.melon.com/chart/")

title_list = driver.find_elements_by_xpath("//div[contains(@class,'ellipsis rank01')]")
singer_list = driver.find_elements_by_xpath("//div[contains(@class,'ellipsis rank02')]")
titleList=list()
nameList=list()
rank=list()
cnt=0
for i in range(0, len(title_list)):
    cnt=cnt+1
    #print("가수:"+singer_list[i].text+", 제목 :"+title_list[i].text)
    if cnt%2 !=0:
        rank.append(cnt)
        titleList.append(title_list[i].text)
        nameList.append(singer_list[i].text)
driver.quit()
data = {'순위': rank,
        '제목': titleList,
        '가수': nameList
        }
df = pd.DataFrame(data,columns=['순위','제목','가수'])

outlook = win32com.client.Dispatch("Outlook.Application")
newMail = outlook.CreateItem(0)
temp = df.to_html(index=False, justify='center')
#메일 to
newMail.To = "sunhee.na@amkor.co.kr"
#메일 cc걸기
#newMail.CC = "XXX.XXX@amkor.co.kr"
#메일 제목
newMail.Subject = "멜론 Top100 중 홀수 순위 입니다."
#메일 내용
newMail.HTMLBody = temp
newMail.Send()
```
