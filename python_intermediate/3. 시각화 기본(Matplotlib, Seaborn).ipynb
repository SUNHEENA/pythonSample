## 🎯 matplotlib 이란?

- 파이썬으로 기본적인 차트들을 쉽게 그릴 수 있도록 도와주는 가장 유명한 데이터 시각화 라이브러리
- 막대차트, 파이차트, 라인차트 등을 그릴 수 있다.

## 🎯 matplotlib 기본

### 📑 기본적인 그래프 그려보기

- `.plot(x축 데이터, y축 데이터)` 
- `.show()`, `print(그래프)` 👉 그래프를 화면에 출력

```python
!pip install matplotlib
```

```python
#꺾은선 그래프 그리기
import matplotlib.pyplot as plt

x = [1, 2, 3]
y = [4, 10, 6]

plt.plot(x, y)
plt.show()
```

```python
#그래프 출력하기
## 데이터 형태도 같이 출력된다
print(plt.plot(x, y))
```

```python
# 그래프만 출력하기
plt.plot(x,y)
plt.show()
```

### 📑 제목, 축 이름, 범례 추가

- `.title` 👉 그래프 제목
- `.xlabel` 👉 x축 이름
- `.ylabel` 👉 y축 이름
- `.legend` 👉 범례

```python
# 다음 데이터에 제목, 축 이름, 범례를 추가해봅시다.
x = ['Math', 'Korean', 'English']
y = [90, 100, 50]
plt.plot(x,y)
plt.show()
```

```python
# 제목 추가하기
plt.plot(x,y)
plt.title('Midterm')
```

```python
# 축 이름 추가하기
plt.plot(x,y)
plt.xlabel('Subject')
plt.ylabel('Score')
```

```python
# 범례 추가하기
plt.plot(x,y)

x2 = ['Math', 'Korean', 'English']
y2 = [100, 80, 60]
plt.plot(x2, y2)
plt.legend(['INKI','JEEHYUN'])
```

### 📑축

- `xticks([x축])`

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [5, 6, 9, 8]

plt.plot(x,y)
```

### 📑 범례(legend)

- 범례 선언방식
- 범례 위치바꾸기

```python
# 범례 선언
x = [1, 2, 3, 4]
y = [5, 6, 9, 8]

plt.plot(x, y, label ='Type A')
plt.legend()
```

```python
# 범례 선언방식2
x = [1, 2, 3, 4]
y = [5, 6, 9, 8]

plt.plot(x, y),
plt.legend(['TypeA'])
```

```python
# 범례 위치 바꾸기
plt.plot(x, y, label = 'TypeA')

# 👀 matplotlib legend 검색하면 loc 적용가능한 값 참조가능!
# plt.legend(loc = 'upper right')
plt.legend(loc = 'lower right')
```

```python
# 범례 위치 바꾸기2 (좌표)
## 0~1 사이 값을 넣어줘야 한다.
plt.plot(x, y, label = 'TypeA')
plt.legend(loc=(0, 0.5))
```

## 🎯 산점도 (Scatter plot)

- dataframe 형태로 엑셀을 받아와 scatter plot 그려보자

```python
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False

import pandas as pd
```

```python
import pandas as pd
df = pd.read_csv("data/score.csv")
df
```

```python
plt.scatter(df['english'], df['korean'])
plt.xlabel('영어 점수')
plt.ylabel('수학 점수')
plt.show()
```

### 📑 사이즈

```python
# sizes 옵션을 통해서 사이즈 조절
## 나이에 따라 점의 크기가 달라지도록 해보자
sizes = df['age'] * 10
plt.scatter(df['english'], df['korean'], sizes = sizes)
plt.xlabel('영어 점수')
plt.ylabel('수학 점수')
```

### 📑 색깔

```python
plt.scatter(df['english'], df['korean'], sizes = sizes, c=df['class'])
plt.xlabel('영어 점수')
plt.ylabel('수학 점수')
plt.colorbar(ticks=[1,2], label='반')
```

### 📑 판다스로 데이터 불러온 후, seaborn 불러오기

```python
import pandas as pd

df = pd.read_csv('data/score.csv')
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

### 📑 'name'과 'math' 요소를 기준으로 산포도 그리기

```python
sns.relplot(data=df, x='name', y = 'math')
```

### 📑 범주 데이터 넣기 & 사이즈 조절

- `hue` 👉 뭐를 기준으로 구분할 건지

## 🎯 그래프 종류

```python
df
```

```python
# sns.relplot(data=df, x='name', y='math')
df
sns.relplot(data=df, x='name', y='math', hue='class', aspect=3)
```

### 📑Scatter Plot

- 👀 https://seaborn.pydata.org/generated/seaborn.scatterplot.html

```python
sns.catplot(data = df, x='class', y='science', aspect=2)
sns.catplot(data = df, x='class', y='math', jitter=False , aspect=2)
sns.catplot(data = df, x='class', y='english', aspect=2, hue='sex')
```

### 📑 Box Plot

- 👀 https://seaborn.pydata.org/generated/seaborn.boxplot.html



```python
# sns.catplot(data=df, x='class', y='math', kind='box')
# sns.catplot(data=df, x='class', y='math', kind='box', hue='sex')

# sns.boxplot(data=df, x='class', y='math')
sns.boxplot(data=df, x='class', y='math', hue='sex')
```

### 📑 Pair plot

- 👀 https://seaborn.pydata.org/generated/seaborn.pairplot.html

```python
sns.pairplot(data=df, hue='sex', x_vars=['class','math'], 
             y_vars=['age', 'math'], height=2)
```

### 📑 Bar Plot

- `sns.load_dataset` 함수는 샘플데이터를 제공한다 
- 👀 https://seaborn.pydata.org/generated/seaborn.load_dataset.html

```python
import matplotlib.pyplot as plt
import seaborn as sns
titanic = sns.load_dataset("titanic")
titanic
# titanic['survived']
```

```python
# barplot - 가장 자주 쓰이는 형태중 하나
sns.catplot(x="sex",y="survived", hue="class", kind="bar",data=titanic)
```

### 📑 Count Plot

- 항목별 갯수 세주는 plot
- 👀 https://seaborn.pydata.org/generated/seaborn.countplot.html

```python
ax = sns.countplot(x="class", data=titanic, hue='who')
```

```python
#그래프 가로로 그리기
sns.countplot(y="class", data=titanic, hue='who')
```

```python
# catplot에서 kind 속성에 "count"를 줘서도 그리기가 가능하다.
sns.catplot(x="class", hue="who", col="survived", data=titanic, kind="count", height=4)
```
