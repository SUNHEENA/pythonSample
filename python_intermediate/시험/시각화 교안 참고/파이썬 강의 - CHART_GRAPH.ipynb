```python
subject=['국어','영어','수학','과학']
subject
```

```python
df['총점'] = df[subject].sum(axis=1) # 총점 column 추가,axis=1 -> 열방향으로의 data 처리
df['평균'] = df['총점']/len(subject) # 퍙군 column 추가 총점/과목수
df
```

```python
# 과목별 합산 행 추가
df.loc['Total',:]=df[['국어','영어','수학','과학']].sum()
df
```

```python
# NaN 값에 value 채우기
df=df.fillna(value='')
df
```

```python
#행 삭제 
df=df.drop(['Total'])
df
```

```python

# 1반과 2반의 평균 구하기

# 반이 1 인 것만 True 를 리턴하고 해당 data 만 class1 data frame 에 저장
class1 = df[df['반'] == 1]
class2 = df[df['반'] == 2]

avgOfClass1 = class1['총점'].sum() / (6*4)
avgOfClass2 = class2['총점'].sum() / (6*4)

print(avgOfClass1,avgOfClass2)

avgOfClass1 = class1['평균'].sum() / (6)
avgOfClass2 = class2['평균'].sum() / (6)

print(avgOfClass1,avgOfClass2)
```

```python
# heatmap
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['figure.figsize'] = [10, 8]
flights = sns.load_dataset('flights') 
# seaborn에 내장되어 있는 dataFrame
#연도와 월별 비행기 승객의 숫자
flights.head()
```

```python
df = flights.pivot('month', 'year', 'passengers')
df
```

```python

# heatmap by seaborn
ax = sns.heatmap(df)
plt.title('Heatmap of Flight by seaborn', fontsize=20)
plt.show() 

```

```python

# annotate each cell with the numeric value of integer format
#annot=True argument를 써서 각 셀에 숫자를 입력(annotate each cell with numeric value)
#fmt='d' 는 정수 형태(integer format)로 숫자를 입력하라는 option

sns.heatmap(df, annot=True, fmt='d')
plt.title('Annoteat cell with numeric value', fontsize=20)
plt.show()

```
