```python
import pandas as pd
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
import numpy as np
import xlwt
import win32com.client
import lxml
```

```python
def makeDFPart(temp):
    resultList = pd.DataFrame({'이름': temp['성명(*)'],'근무지':temp['근무지'],'입사일':temp['입사일']})
    for i in resultList.index:
        if str(resultList["입사일"][i]) != 'nan':
            resultList["입사일"][i] = str(resultList["입사일"][i])[0:4]
    return resultList
```

```python
def send_mail(to, subject, content,df):
    style = """ .t2{font-size:10pt; font-family:"맑은 고딕"}
                  .t3{font-size:11pt; font-family:"맑은 고딕"}
                  .dataframe  {font-size: 11pt; font-family: Arial; border: 1px solid; border-color:black; border-collapse: collapse;text-align:center;}
                  .dataframe th {padding: 5px; background-color:rgb(204,217,255); border-color:black; border-collapse: collapse;}
                  .dataframe  td {padding: 5px; border-color:black; border-collapse: collapse;}"""
    temp = df.to_html(index=False, justify='center')
    content = """
              <html>
                  <head><style> {0} </style></head>
                  <body>
                      <div class="t2"> 
                      안녕하세요. <br><br>
                      입사일 정보 보내 드립니다.<br><br> 
                      {1}<br>{2}
                    </div>""".format(style, content, temp)
    # Outlook Object Model 불러오기
    new_Mail = win32com.client.Dispatch("Outlook.Application").CreateItem(0)
    # 메일 수신자
    new_Mail.To = to
    # 메일 참조
    # new_Mail.CC = "mail-add-for-cc@testadd.com"
    # 메일 제목
    new_Mail.Subject = subject
    # 메일 내용
    new_Mail.HTMLBody = content
    # 첨부파일 추가
    # if atch:
    #     for file in atch:
    #         new_Mail.Attachments.Add(file)
    # 메일 발송
    new_Mail.Send()
```

```python
def chromeDriver():
    options = webdriver.ChromeOptions()
    #     options.add_argument('headless')  # 실행하는 것을 안보이게 함
    options.add_argument('window-size=1920x1080')
    options.add_argument("disable-gpu")
    options.add_experimental_option("prefs", {
        #"download.default_directory": path,
        "download.prompt_for_download": False,
        "download.directory_upgrade": True,
        "safebrowsing.enabled": True,
        "profile.default_content_setting_values.automatic_downloads": 1
    })
    # 크롬드라이버 최신 설치
    driver = webdriver.Chrome(executable_path=ChromeDriverManager().install(), options=options)

    return driver

driver = chromeDriver()
driver.get('C://판다스실습//name_list.html')
resultSet = driver.find_elements_by_class_name("dataframe") # 전체 data
for temp in resultSet:
    if temp.text != "":
        r = temp.text
fixData = r.split('\n')
count=0
titleList=list()
resultList=list()
for dix in fixData:
    x = dix.split(' ')
    if count==0:
        for z in x:
            titleList.append(z)
    else :
        resultList.append(x)
    count += 1
driver.close()
result = pd.DataFrame(resultList, columns=[titleList])
result

#result = pd.read_html('C://판다스실습//name_list.html',encoding='UTF-8')
```
